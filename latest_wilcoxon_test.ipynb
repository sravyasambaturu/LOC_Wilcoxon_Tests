```python
from google.colab import files
import pandas as pd
import io


uploaded = files.upload()


for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))


  df = pd.read_excel(io.BytesIO(uploaded[fn]), sheet_name='Wilcoxon - loc 5.13 vs 7.5')


df
```

```python
from scipy.stats import wilcoxon


column_a = df['Version_7.5']
column_b = df['Version_5.13']


statistic, p_value = wilcoxon(column_a, column_b)


print("Wilcoxon signed-rank test:")
print("Statistic:", statistic)
print("P-value:", p_value)


alpha = 0.05
if p_value < alpha:
  print("The difference between the two columns is statistically significant.")
else:
  print("The difference between the two columns is not statistically significant.")
```
